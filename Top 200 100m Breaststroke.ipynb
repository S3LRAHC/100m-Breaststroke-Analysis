```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('100m_breaststroke.csv')
```

```python
df
```

```python
meetNames = df.loc[:, 'meet_name']
```

```python
meetNames
```

```python
olympicCount = 0
meetCount = len(meetNames)
for i in range(len(meetNames)):
    if 'olympic' in meetNames.iloc[i].casefold() and 'trial' not in meetNames.iloc[i].casefold():
        olympicCount += 1        
olympicProportion = olympicCount / meetCount
print(olympicProportion)
```
