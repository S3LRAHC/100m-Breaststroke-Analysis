```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('100m_breaststroke.csv')
```

```python
df
```

```python
df.meet_name
```

```python
olympicCount = 0
meetCount = len(df.meet_name)
for i in range(meetCount):
    if 'olympic' in df.meet_name.iloc[i].casefold() and 'trial' not in df.meet_name.iloc[i].casefold():
        olympicCount += 1        
olympicProportion = olympicCount / meetCount
print(olympicProportion)
```

```python
df.height = df.height.replace('-', '')
```

```python
df.height = df.height.replace('', -1)
```

```python
df.height = df.height.astype('int')
```

```python
df
```

```python
plt.scatter(df.height, df.fina_points)
plt.xlim([175,200])
plt.xlabel('Height (cm)')
plt.ylabel('Fina Points')
plt.title('Heights of Top 100m Breaststrokers')
plt.show()
```

```python
df.height.corr(df.fina_points)
```

```python
# make heat map of this? seems more dense near age 20 

plt.scatter(df.age_at_meet, df.fina_points)
plt.xlabel('Age')
plt.ylabel('Fina Points')
plt.title('Ages of Top 100m Breaststrokers')
plt.show()
```

```python
df.age_at_meet.corr(df.fina_points)
```

```python
# Lots of top 200 performances occurred when competition city had same time zone as home city.
# should probably make this a histogram somehow

plt.scatter(df.time_difference_hours, df.fina_points)
plt.xlabel('Time Difference of Meet City and Home (hours)')
plt.ylabel('Fina Points')
plt.title('Effect of Jet Lag on Top 100m Breaststrokers')
plt.show()
```
