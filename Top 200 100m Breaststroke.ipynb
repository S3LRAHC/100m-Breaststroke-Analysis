# Importing Libraries & .csv File

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('100m_breaststroke.csv')
```

```python
df
```

# Finding Locations of Meets

```python
olympicCount = 0
nationalCount = 0
worldsCount = 0
trialsCount = 0
meetCount = len(df.meet_name)
otherCount = 0
otherList = []
for i in range(meetCount):
    meetName = df.meet_name.iloc[i].casefold()
    if 'trial' not in meetName and 'trials' not in meetName:
        if 'olympic' in meetName:
            olympicCount += 1  
        elif 'national' in meetName:
            nationalCount += 1
        elif 'fina' in meetName:
            worldsCount += 1
        else:
            otherCount += 1
            otherList.append(meetName)
    elif 'trial' in meetName or 'trials' in meetName:
        trialsCount += 1
```

```python
print(olympicCount)
print(nationalCount)
print(worldsCount)
print(trialsCount)
print(otherCount)
```

# Analyzing Height of Swimmers

```python
df.height = df.height.replace('-', -1)
```

```python
df.height = df.height.astype('int')
```

```python
plt.scatter(df.height, df.fina_points, color='b')
plt.xlim([175,200])
plt.xlabel('Height (cm)')
plt.ylabel('Fina Points')
plt.title('Heights of Top 100m Breaststrokers')
plt.show()
```

```python
plt.hist(df.height, bins=[175, 180, 185, 190, 195, 200], rwidth=0.95, color='b')
plt.ylim([0,25])
plt.xlim([175,200])
plt.xlabel('Height (cm)')
plt.ylabel('Number of athletes')
plt.title('Heights of Top 100m Breaststrokers')
```

```python
# finding correlation between height and amount of fina points scored

df.height.corr(df.fina_points)
```

```python
heights = []
for i in range(len(df)):
    if df.height.iloc[i] > -1:
        heights.append(df.height.iloc[i])
heightDf = pd.Series(heights)
```

```python
# summary of height data

heightDf.describe()
```

```python
# finding percent of swimmers between 185-195cm tall

heightRange = 0
for i in range(len(heightDf)):
    if heightDf.iloc[i] >= 185 and heightDf.iloc[i] <= 195:
        heightRange += 1
        
percentRange = 100 * (heightRange / len(heightDf))
print(percentRange)
```

# Analyzing Ages of Swimmers

```python
plt.scatter(df.age_at_meet, df.fina_points, color='c')
plt.xlabel('Age')
plt.ylabel('Fina Points')
plt.title('Ages of Top 100m Breaststrokers')
plt.show()
```

```python
plt.hist(df.age_at_meet, rwidth=0.95, color='c')
plt.xlabel('Age')
plt.ylabel('Number of athletes')
plt.title('Ages of Top 100m Breaststrokers')
```

```python
# finding correlation between age and amount of fina points scored

df.age_at_meet.corr(df.fina_points)
```

```python
# summary of age data

df.age_at_meet.describe()
```

```python
# percentage of swimmers who are between 20-26 years old

ageRange = 0
for i in range(len(df)):
    if df.age_at_meet.iloc[i] >= 20 and df.age_at_meet.iloc[i] <= 26:
        ageRange += 1
        
percentRange = 100 * (ageRange / len(df))
print(percentRange)
```

# Time Zone & Location Visualizations

```python
plt.hist(df.time_difference_hours, rwidth=0.95, color='green', bins=10)
plt.xlabel('Time Difference of Meet City and Home (hours)')
plt.ylabel('Number of athletes')
plt.title('Effect of Jet Lag on Top 100m Breaststrokers')
```

```python
# finding percentage of swimmers who swam their best with a +- 1 hour time zone difference to home

noDiff = 0
for i in range(len(df)):
    if df.time_difference_hours.iloc[i] >= -1 and df.time_difference_hours.iloc[i] <= 1:
        noDiff += 1
    
percent = 100 * (noDiff / len(df))
print(percent)
```

```python
pieValues = [olympicCount, nationalCount, worldsCount, trialsCount, otherCount]
pieLabels = ['Olympics', 'Nationals', 'FINA World Championships', 'Trials', 'Other']
plt.axis('equal')
plt.pie(pieValues, labels=pieLabels, autopct='%0.1f%%')
plt.title('Where Top 100m Breaststroke Performances Took Place')
plt.show()

```
